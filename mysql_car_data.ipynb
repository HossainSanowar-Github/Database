```python
def load_module():
    import mysql.connector as cnn
    login=cnn.connect(host='localhost',user='root',passwd='root@1234')
    return login

mydb=load_module()
```

# Create table and load data into table

```python
def car_database():
    try:
        cursor=mydb.cursor()
        #cursor.execute('drop procedure if exists car_dataset')
        cursor.execute('create database cardatabase')
        cursor.execute('create table cardatabase.car(buying VARCHAR(50), maint VARCHAR(50),doors VARCHAR(10),persons VARCHAR(10),lug_boot VARCHAR(50), safety VARCHAR(50))')
    
    except Exception as e:
        print(e)

car_database()
```

# Car Data load into table

```python
cursor=mydb.cursor()
import csv
with open('car.data','r') as f:
    car_data=csv.reader(f,delimiter='\n')
    for row in car_data:
        #print(row)
        #print(type(row))
        #print(len(row))
        lst=row[0].split(',')
        cursor.execute("insert into cardatabase.car values('{0}','{1}','{2}','{3}','{4}','{5}')".format(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5]))

mydb.commit()
```
