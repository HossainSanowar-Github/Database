```python
1. Id number: 1 to 214
2. RI: refractive index
3. Na: Sodium (unit measurement: weight percent in corresponding oxide, as are attributes 4-10)
4. Mg: Magnesium
5. Al: Aluminum
6. Si: Silicon
7. K: Potassium
8. Ca: Calcium
9. Ba: Barium
10. Fe: Iron
11. Type of glass: (class attribute)
```

```python
#Create glassdata_base
# check the database current status
try:
    mydb=conn.connect(host='localhost',user='root',passwd='root@1234')
    #check if the connection is established
    
    query="SHOW DATABASES"
    
    cursor=mydb.cursor()#create a cursor to execute queries
    cursor.execute(query)
    cursor.execute('create database glassdatabase')#create database
    cursor.execute("CREATE table glassdatabase.glass(Id_number INT(10), RI FLOAT(10,5), Na FLOAT(10,5), MG FLOAT(10,5), AI FLOAT(10,5), Si FLOAT(10,5), K FLOAT(10,5), Ca FLOAT(10,5), Ba FLOAT(10,5), Fe FLOAT(10,5), glass_type INT(10))")

    print(cursor.fetchall())
    
except Exception as e:
    mydb.close()
    print(str(e))
```

```python
mydb=conn.connect(host='localhost',user='root',passwd='root@1234')
cursor=mydb.cursor()#create a cursor to execute queries


with open('glass.data','r') as f:
    glass_data=csv.reader(f,delimiter='\n')
    print(glass_data)
    for i in glass_data:
        #print(type(i))
        #print(i)
        print(f'insert into glassdata values({str(i[0])})')
        cursor.execute('insert into glassdatabase.glass values({data})'.format(data=i[0]))
        #cursor.execute('insert into glassdata values({values})'.format(values=((i[0]))))

mydb.commit()
```
